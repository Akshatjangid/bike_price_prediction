```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Used_Bikes.csv")
df
```

```python
df.info()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.isnull().sum()
df.dropna(inplace=True)
```

```python
df.info()
```

```python
## cat col ---- num col
cat_col = df.select_dtypes(include=['object'])
cat_col.head()
```

```python
# num_col = df.select_dtypes(include=['int64', 'float64'])
num_col = df.select_dtypes(exclude="object")
num_col.head()
```

```python
cat_col=cat_col.drop(columns=["bike_name","city"],axis=1)
cat_col.head()
```

```python
# categorical columns --> change to numerical 
## label encoding
# one hot encoding
```

```python
# one hot encoding
pd.get_dummies(cat_col).astype(int).head()                    # astype is use to convewrt the datatype to numeric
```

```python
cat_col.head()
```

```python
## label encoding
cat_col["owner"].value_counts()

```

```python
dt={"First Owner":1,
    "Second Owner":2,
    "Third Owner":3,
    "Fourth Owner Or More":4}
```

```python
cat_col["owner"]=cat_col["owner"].map(dt)                     # when we doublerun the code it will  change the value as NaN

```

```python
cat_col.head()
```

```python
cat_col.brand.value_counts()
```

```python
dt1={i:key for key ,i in enumerate(list(cat_col.brand.unique()))}                   # it is use to convert data into dictonary
```

```python
dt1
```

```python
cat_col.brand=cat_col.brand.map(dt1)
```

```python
cat_col.head(20)
```

```python
# merge categorical and numerical columns
final_df=pd.concat([num_col,cat_col],axis=1)             # to merge col and num
```

```python
final_df.head()
```

```python
# now install scikit-learn
## model traning --> traning(20%) and testing(20%)
```

```python
matrix = final_df.corr()
matrix                                # its tell that -ive data is very low and +ive data is very high and it is very good for model traning
```

```python
sns.heatmap(matrix,annot=True,cmap='coolwarm',linewidths=0.5)
```

```python
sns.heatmap(matrix,annot=True,cmap='coolwarm')
```

```python
## divide the data into x and y or independent and dependent variable
x = final_df.drop("price",axis=1)
y= final_df["price"]
```

```python
## trainning and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                                test_size=0.2,
                                               random_state=42)             # test_size= 0.2 mean 20% of data is used for testing and 80% is used for traning
            # random_state is used to get the same data every time we run the code means 
                                                                     #it will not change the data every time we run the code
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)  # fit the model with traning data
```

```python
lr.score(x_train,y_train)  # score the model with traning data
```

```python

lr.score(x_test,y_test)    # score the model with testing data
```

```python
sns.scatterplot(x=y_test,y=lr.predict(x_test))  # scatter plot to see the relation between actual and predicted values
```

```python
# model evaluation
y_pred=lr.predict(x_test)  # predict the values using testing data
y_pred
```

```python
## making a dataframe 
df1=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})  # create a dataframe with actual and predicted values
df1.head(5)  # show the first 20 rows of the dataframe
```

```python
# model evaluation metrics
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y_test,y_pred))  # mean absolute error
print("MSE:",metrics.mean_squared_error(y_test,y_pred))  # mean squared error
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))  # root mean squared error
```

```python
sns.scatterplot(x=y_test,y=lr.predict(x_test))  # scatter plot to see the relation between actual and predicted values
plt.xlabel("Actual Price")  # x label
plt.ylabel("Predicted Price")  # y label
plt.title("Actual vs Predicted Price")  # title of the plot
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # line for actual vs predicted
plt.show()  # show the plot
```
